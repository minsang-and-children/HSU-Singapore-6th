```python
import pandas as pd
import numpy as np
import os
from pathlib import Path

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
```

```python
def check_dataframe(df, filename):
    """Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò ÌíàÏßàÏùÑ Ï≤¥ÌÅ¨ÌïòÎäî Ìï®Ïàò"""
    print(f"\n{'='*80}")
    print(f"üìä ÌååÏùº: {filename}")
    print(f"{'='*80}")
    
    # 1. Í∏∞Î≥∏ Ï†ïÎ≥¥
    print(f"\n1Ô∏è‚É£ Í∏∞Î≥∏ Ï†ïÎ≥¥")
    print(f"   - Ìñâ Í∞úÏàò: {len(df):,}")
    print(f"   - Ïó¥ Í∞úÏàò: {len(df.columns)}")
    print(f"   - Î©îÎ™®Î¶¨ ÏÇ¨Ïö©Îüâ: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    
    # 2. Ïª¨Îüº Ï†ïÎ≥¥
    print(f"\n2Ô∏è‚É£ Ïª¨Îüº Ï†ïÎ≥¥")
    print(df.dtypes)
    
    # 3. Í≤∞Ï∏°Ïπò Ï≤¥ÌÅ¨
    print(f"\n3Ô∏è‚É£ Í≤∞Ï∏°Ïπò (Missing Values)")
    missing = df.isnull().sum()
    missing_pct = (missing / len(df) * 100).round(2)
    missing_df = pd.DataFrame({
        'Í≤∞Ï∏°Ïπò Í∞úÏàò': missing,
        'ÎπÑÏú®(%)': missing_pct
    })
    missing_df = missing_df[missing_df['Í≤∞Ï∏°Ïπò Í∞úÏàò'] > 0]
    
    if len(missing_df) > 0:
        print(missing_df)
    else:
        print("   ‚úÖ Í≤∞Ï∏°Ïπò ÏóÜÏùå!")
    
    # 4. Ï§ëÎ≥µ Îç∞Ïù¥ÌÑ∞
    duplicates = df.duplicated().sum()
    print(f"\n4Ô∏è‚É£ Ï§ëÎ≥µ Ìñâ")
    print(f"   - Ï§ëÎ≥µ Ìñâ Í∞úÏàò: {duplicates:,}")
    if duplicates > 0:
        print(f"   ‚ö†Ô∏è Ï§ëÎ≥µ Îç∞Ïù¥ÌÑ∞ Î∞úÍ≤¨!")
    else:
        print(f"   ‚úÖ Ï§ëÎ≥µ ÏóÜÏùå!")
    
    # 5. ÎÇ†Ïßú Ïª¨Îüº Ï≤¥ÌÅ¨ (ÏûàÎäî Í≤ΩÏö∞)
    date_cols = [col for col in df.columns if 'date' in col.lower() or 'time' in col.lower()]
    if date_cols:
        print(f"\n5Ô∏è‚É£ ÎÇ†Ïßú Î≤îÏúÑ")
        for col in date_cols:
            try:
                df[col] = pd.to_datetime(df[col])
                print(f"   - {col}:")
                print(f"     ÏãúÏûë: {df[col].min()}")
                print(f"     Ï¢ÖÎ£å: {df[col].max()}")
                print(f"     Í∏∞Í∞Ñ: {(df[col].max() - df[col].min()).days} Ïùº")
            except:
                print(f"   - {col}: ÎÇ†Ïßú Î≥ÄÌôò Ïã§Ìå®")
    
    # 6. Ïà´ÏûêÌòï Ïª¨Îüº Í∏∞Î≥∏ ÌÜµÍ≥Ñ
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        print(f"\n6Ô∏è‚É£ Ïà´ÏûêÌòï Ïª¨Îüº Í∏∞Î≥∏ ÌÜµÍ≥Ñ")
        print(df[numeric_cols].describe().round(2))
    
    return df
```

```python
print("="*80)
print("üîç GICS Îç∞Ïù¥ÌÑ∞ - Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò Î∂ÑÏÑù")
print("="*80)

sectors_path = os.path.join(base_path, 'data', 'gics', 'sector.csv')
industry_groups_path = os.path.join(base_path, 'data', 'gics', 'industry_group.csv')
industries_path = os.path.join(base_path, 'data', 'gics', 'industry.csv')
sub_industries_path = os.path.join(base_path, 'data', 'gics', 'sub_industry.csv')

gics_files = [sectors_path, industry_groups_path, industries_path, sub_industries_path]

for filename in gics_files:
    print(f"\nüìÅ {filename}")
    df = pd.read_csv(filename)
    
    # symbol Ïª¨ÎüºÏù¥ ÏûàÎäîÏßÄ ÌôïÏù∏
    if 'symbol' in df.columns:
        # Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò
        missing_by_symbol = df.groupby('symbol').apply(
            lambda x: pd.Series({
                'Í≤∞Ï∏°Ïπò_Í∞úÏàò': x.isnull().sum().sum(),
                'Ï†ÑÏ≤¥_ÌñâÏàò': len(x)
            })
        ).sort_values('Í≤∞Ï∏°Ïπò_Í∞úÏàò', ascending=False)
        
        missing_symbols = missing_by_symbol[missing_by_symbol['Í≤∞Ï∏°Ïπò_Í∞úÏàò'] > 0]
        
        if len(missing_symbols) > 0:
            print(f"  ‚ö†Ô∏è Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî Ï¢ÖÎ™©: {len(missing_symbols)}Í∞ú")
            print(f"\n  ÏÉÅÏúÑ 10Í∞ú:")
            print(missing_symbols.head(10))
        else:
            print(f"  ‚úÖ Î™®Îì† Ï¢ÖÎ™©Ïóê Í≤∞Ï∏°Ïπò ÏóÜÏùå")
    else:
        print(f"  ‚ÑπÔ∏è symbol Ïª¨Îüº ÏóÜÏùå - Ï†ÑÏ≤¥ Í≤∞Ï∏°ÏπòÎßå ÌëúÏãú")
        missing = df.isnull().sum()
        if missing.sum() > 0:
            print(missing[missing > 0])
        else:
            print("  ‚úÖ Í≤∞Ï∏°Ïπò ÏóÜÏùå")
```

### 3. ÏùºÎ≥Ñ Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ - Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò


```python
print("="*80)
print("üîç ÏùºÎ≥Ñ Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ - Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò Î∂ÑÏÑù")
print("="*80)

price_path = os.path.join(base_path, 'data', 'price')
open_path = os.path.join(price_path, 'open.csv')
high_path = os.path.join(price_path, 'high.csv')
low_path = os.path.join(price_path, 'low.csv')
close_path = os.path.join(price_path, 'close.csv')
volume_path = os.path.join(price_path, 'volume.csv')
vwap_path = os.path.join(price_path, 'vwap.csv')

price_files = [open_path, high_path, low_path, close_path, volume_path, vwap_path]

for filename in price_files:
    print(f"\nüìÅ price/{filename}")
    df = pd.read_csv(filename)
    
    # ÎÇ†Ïßú Ïª¨Îüº Ï†úÏô∏ÌïòÍ≥† Í∞Å Ï¢ÖÎ™©(Ïª¨Îüº)Î≥Ñ Í≤∞Ï∏°Ïπò Ï≤¥ÌÅ¨
    # Î≥¥ÌÜµ Ï≤´ Ïª¨ÎüºÏùÄ ÎÇ†Ïßú, ÎÇòÎ®∏ÏßÄÎäî Í∞Å Ï¢ÖÎ™©Ïùò Îç∞Ïù¥ÌÑ∞
    date_col = df.columns[0]
    symbol_cols = df.columns[1:]
    
    # Í∞Å Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò
    missing_by_symbol = df[symbol_cols].isnull().sum().sort_values(ascending=False)
    
    # Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî Ï¢ÖÎ™©Îì§
    missing_symbols = missing_by_symbol[missing_by_symbol > 0]
    
    if len(missing_symbols) > 0:
        print(f"  ‚ö†Ô∏è Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî Ï¢ÖÎ™©: {len(missing_symbols)}Í∞ú")
        print(f"  Ï¥ù Í≤∞Ï∏°Ïπò: {missing_symbols.sum()}Í∞ú")
        print(f"\n  Í≤∞Ï∏°Ïπò ÏÉÅÏúÑ 10Í∞ú Ï¢ÖÎ™©:")
        for symbol, count in missing_symbols.head(10).items():
            print(f"    {symbol}: {count}Í∞ú ({count/len(df)*100:.1f}%)")
    else:
        print(f"  ‚úÖ Î™®Îì† Ï¢ÖÎ™©Ïóê Í≤∞Ï∏°Ïπò ÏóÜÏùå (Ï¢ÖÎ™© Ïàò: {len(symbol_cols)}Í∞ú)")
```

### 5. Î∂ÑÎ¥â Îç∞Ïù¥ÌÑ∞ - Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò ÏÉÅÏÑ∏ Î∂ÑÏÑù


```python
print("="*80)
print("üîç Î∂ÑÎ¥â Îç∞Ïù¥ÌÑ∞ - Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò Î∂ÑÏÑù (ÏÉòÌîå)")
print("="*80)

sample_files = [
    'close_0900_0910.csv',  # Ïû• ÏãúÏûë
    'close_1200_1210.csv',  # Ï†êÏã¨ ÏãúÍ∞Ñ
    'close_1520_1530.csv',  # Ïû• ÎßàÍ∞ê
    'open_0900_0910.csv',   # ÏãúÍ∞Ä ÏÉòÌîå
]

price_minutely_path = os.path.join(base_path, 'data', 'price_minutely')

for filename in sample_files:
    print(os.path.join(price_minutely_path, filename))
    df = pd.read_csv(os.path.join(price_minutely_path, filename))
    
    # ÎÇ†Ïßú/Ïù∏Îç±Ïä§ Ïª¨Îüº Ï†úÏô∏ÌïòÍ≥† Ï¢ÖÎ™© Ïª¨ÎüºÎì§Îßå
    date_col = df.columns[0]
    symbol_cols = df.columns[1:]
    
    print(f"  Ï¥ù Ìñâ Ïàò: {len(df)}, Ï¥ù Ï¢ÖÎ™© Ïàò: {len(symbol_cols)}")
    
    # Í∞Å Ï¢ÖÎ™©Î≥Ñ Í≤∞Ï∏°Ïπò Í≥ÑÏÇ∞
    missing_by_symbol = df[symbol_cols].isnull().sum()
    missing_pct = (missing_by_symbol / len(df) * 100).round(2)
    
    # Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî Ï¢ÖÎ™©Îßå ÌïÑÌÑ∞ÎßÅ
    has_missing = missing_by_symbol[missing_by_symbol > 0]
    
    if len(has_missing) > 0:
        print(f"  ‚ö†Ô∏è Í≤∞Ï∏°ÏπòÍ∞Ä ÏûàÎäî Ï¢ÖÎ™©: {len(has_missing)}Í∞ú")
        print(f"  Ï¥ù Í≤∞Ï∏°Ïπò: {has_missing.sum()}Í∞ú")
        
        # Í≤∞Ï∏°Ïπò Ï†ïÎ≥¥Î•º DataFrameÏúºÎ°ú Ï†ïÎ¶¨
        missing_info = pd.DataFrame({
            'Ï¢ÖÎ™©': has_missing.index,
            'Í≤∞Ï∏°Ïπò_Í∞úÏàò': has_missing.values,
            'ÎπÑÏú®(%)': [missing_pct[s] for s in has_missing.index]
        }).sort_values('ÎπÑÏú®(%)', ascending=False)
        
        print(f"\n  Í≤∞Ï∏°Ïπò ÏÉÅÏúÑ 10Í∞ú Ï¢ÖÎ™©:")
        for idx, row in missing_info.head(10).iterrows():
            print(f"    {row['Ï¢ÖÎ™©']}: {row['Í≤∞Ï∏°Ïπò_Í∞úÏàò']:.0f}Í∞ú ({row['ÎπÑÏú®(%)']:.1f}%)")
        
        # Í≤∞Ï∏°Ïπò ÎπÑÏú® Î∂ÑÌè¨
        print(f"\n  üìä Í≤∞Ï∏°Ïπò ÎπÑÏú® Î∂ÑÌè¨:")
        print(f"    - 1% ÎØ∏Îßå: {len(missing_info[missing_info['ÎπÑÏú®(%)'] < 1])}Í∞ú")
        print(f"    - 1-5%: {len(missing_info[(missing_info['ÎπÑÏú®(%)'] >= 1) & (missing_info['ÎπÑÏú®(%)'] < 5)])}Í∞ú")
        print(f"    - 5-10%: {len(missing_info[(missing_info['ÎπÑÏú®(%)'] >= 5) & (missing_info['ÎπÑÏú®(%)'] < 10)])}Í∞ú")
        print(f"    - 10% Ïù¥ÏÉÅ: {len(missing_info[missing_info['ÎπÑÏú®(%)'] >= 10])}Í∞ú")
    else:
        print(f"  ‚úÖ Î™®Îì† Ï¢ÖÎ™© Í≤∞Ï∏°Ïπò ÏóÜÏùå")
    
    print("-"*80)
```
